* ê°„ëµíˆ ë§í•´ì„œ, ì‚¬ê±´ $E$ì™€ $F$ê°€ ë…ë¦½ì´ë¼ëŠ” ê²ƒì€, ìš°ë¦¬ê°€ $E$ì˜ ì‚¬ê±´ì´ ë°œìƒí–ˆëŠ”ì§€ì— ëŒ€í•œ ì •ë³´ë¥¼ ì•Œê³  ìˆì„ ë•Œ, ì´ê²ƒì´ $F$ì— ëŒ€í•œ ì •ë³´ë¥¼ ì£¼ì§€ ëª»í•˜ë©´ ì‚¬ê±´ $E$ì™€ $F$ëŠ” ë…ë¦½ì´ë¼ í•œë‹¤.

## ì¡°ê±´ë¶€ í™•ë¥ 

* ë§Œì•½ ì‚¬ê±´ $E$ì™€ $F$ê°€ ë…ë¦½ì´ ì•„ë‹ ìˆ˜ ìˆìœ¼ë©´, ë‹¤ìŒê³¼ ê°™ì´ ì¡°ê±´ë¶€ í™•ë¥ ì„ ì •ì˜

$$
P(E|F) = \frac{P(E \cap F)}{P(F)}
$$

* ì‚¬ê±´ $F$ê°€ ë°œìƒí•˜ì˜€ì„ ë•Œ, $E$ê°€ ë°œìƒí•  í™•ë¥ 


### Example

* ë‘ ìë…€ê°€ ìˆëŠ” ê°€ì¡±ì„ ìƒê°í•´ ë³´ì.

* ì²«ì§¸ê°€ ë”¸(G)ì¼ ë•Œ, ë‘ ìë…€ê°€ ëª¨ë‘ ë”¸(B)ì¼ í™•ë¥ ?

$$
P(B|G) = \frac{P(B \cap G)}{P(G)} = \frac{P(B)}{P(G)} = \frac{1}{2}
$$

* ì ì–´ë„ í•˜ë‚˜ì˜ ì•„ì´ê°€ ë”¸ì¼ ë•Œ(L), ë‘ ì•„ì´ê°€ ëª¨ë‘ ë”¸(B)ì¼ í™•ë¥ ?

$$P(B|L) = \frac{P(B \cap L)}{P(L)} = \frac{P(B)}{P(L)} = \frac{1}{3}$$


### ëœë¤ ê°€ì¡±ì„ ë§Œë“¤ì–´ ê²°ê³¼ í™•ì¸

````python
import random

def random_kid(): return random.choice(["boy", "girl"])

both_girls, older_girl, either_girl = 0, 0, 0
random.seed(0)

for _ in range(10000):
    younger = random_kid()
    older = random_kid()
    
    if older == "girl": older_girl += 1
    if older == "girl" and younger == "girl": both_girls += 1
    if older == "girl" or younger == "girl": either_girl += 1

print("P(both | older):", both_girls / older_girl)      
print("P(both | either): ", both_girls / either_girl)
````

## Bayes's theorem

* $P(F|E)$ë¥¼ ì•Œ ë•Œ $P(E|F)$ë¥¼ ê³„ì‚°
$$ P(E|F) = \frac{P(E \cap F)}{P(F)} = \frac{P(F|E)P(E)}{P(F)} $$

* ì‚¬ê±´ $F$ëŠ” $F \cap E$ì™€ $F \cap E^c$ë¡œ ë‚˜ëˆŒ ìˆ˜ ìˆìŒ
$$P(E|F) = \frac{P(F|E)P(E)}{P(F|E)P(E) + P(F|E^c)P(E^c)} $$

ì„ì˜ì˜ $a$ì™€ $b$ ì‚¬ì´ì˜ ê· ë“± ë¶„í¬ì— ëŒ€í•œ pdfì™€ cdf í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì—¬ ë³´ì.

### ì—°ì† ë¶„í¬ â€“ ì •ê·œ ë¶„í¬

* ë²¨ ëª¨ì–‘ì˜ ì •ê·œë¶„í¬ëŠ” í‰ê·  ğœ‡ì™€ í‘œì¤€ í¸ì°¨ ğœë¥¼ ì´ìš©í•˜ì—¬ í‘œí˜„
* ì •ê·œ ë¶„í¬ì˜ pdf:
$$ f(x) = \frac{1}{\sqrt{2\pi}\sigma} \exp \left( -\frac{ (x-\mu)^2 }{2\sigma^2} \right) $$

````python
import math
def normal_pdf(x, mu=0, sigma=1):
    sqrt_two_pi = math.sqrt(2 * math.pi)
    return (math.exp(-(x - mu) ** 2 / 2 /sigma ** 2)/(sqrt_two_pi * sigma))
````

#### ì •ê·œë¶„í¬ pdf ê·¸ë¦¬ê¸°

````python
%matplotlib inline 
````

````python
from matplotlib import pyplot as plt
import math
xs = [x / 10.0 for x in range(-50, 50)]
plt.plot(xs,[normal_pdf(x,sigma=1) for x in xs], '-', label='mu=0,sigma=1')
plt.plot(xs,[normal_pdf(x,sigma=2) for x in xs], '--', label='mu=0,sigma=2')
plt.plot(xs,[normal_pdf(x,sigma=0.5) for x in xs], ':', label='mu=0,sigma=0.5')
plt.plot(xs,[normal_pdf(x,mu=-1)   for x in xs], '-.', label='mu=-1,sigma=1')
plt.legend()
plt.show()
````

#### ì •ê·œë¶„í¬ì˜ cdf

* ì •ê·œ ë¶„í¬ì˜ cdf:
$$ F(x) = \frac{1}{2} \left\{ 1 + erf\left( \frac{x - \mu}{\sqrt{2} \sigma} \right) \right\}$$
* erfëŠ” ì—ëŸ¬ í•¨ìˆ˜ë¼ê³  ë¶ˆë¦¬ë©° ```math.erf```ë¥¼ í†µí•´ ê³„ì‚°

````python
def normal_cdf(x, mu=0, sigma=1):
    return (1+math.erf((x-mu)/math.sqrt(2)/sigma))/2
````

#### ì •ê·œë¶„í¬ cdf ê·¸ë¦¬ê¸°

````python
xs = [x / 10.0 for x in range(-50, 50)]
plt.plot(xs,[normal_cdf(x,sigma=1) for x in xs], '-', label='mu=0,sigma=1')
plt.plot(xs,[normal_cdf(x,sigma=2) for x in xs], '--', label='mu=0,sigma=2')
plt.plot(xs,[normal_cdf(x,sigma=0.5) for x in xs], ':', label='mu=0,sigma=0.5')
plt.plot(xs,[normal_cdf(x,mu=-1) for x in xs], '-.', label='mu=-1,sigma=1')
plt.legend(loc=4) # bottom right
plt.show()
````

#### ì •ê·œë¶„í¬ cdfì˜ ì—­í•¨ìˆ˜

* ì£¼ì–´ì§„ $ğ‘$ì— ëŒ€í•´, $ğ‘=ğ¹(ğ‘§)$ë¥¼ ë§Œì¡±í•˜ëŠ” $ğ‘§$ë¥¼ ì°¾ëŠ”ë‹¤.
* ì´ì§„ íƒìƒ‰ì„ ì´ìš©í•œ ë°©ë²•
  * ë¨¼ì € íƒìƒ‰ ë„ë©”ì¸ì„ ```(left, right)```ìœ¼ë¡œ ì„¤ì •
  * ì´ ë„ë©”ì¸ì˜ ì¤‘ì•™ê°’ì¸ ```mid```ì—ì„œ ```ğ¹(mid)```ë¥¼ ê³„ì‚°
   * ë§Œì•½ ```ğ¹(mid) < ğ‘``` ì´ë©´ ìƒˆë¡œìš´ íƒìƒ‰ ë„ë©”ì¸ ```(left, right) = (mid, right)```ìœ¼ë¡œ ì¬ì„¤ì •  
   * ë§Œì•½ ```ğ¹(mid) > ğ‘``` ì´ë©´ ìƒˆë¡œìš´ íƒìƒ‰ ë„ë©”ì¸ ```(left, right) = (left, mid)```ìœ¼ë¡œ ì¬ì„¤ì •
   * ë§Œì•½ ```ğ¹(mid) â‰ˆ ğ‘``` ì´ë©´ íƒìƒ‰ ì¢…ë£Œ


````python
def inverse_normal_cdf(p, mu=0, sigma=1, tolerance=0.000001):
    # í‘œì¤€ ì •ê·œ ë¶„í¬ê°€ ì•„ë‹Œ ê²½ìš° X = mu + sigma * Zë¥¼ ì´ìš©
    if mu != 0 or sigma != 1:
        return mu + sigma * inverse_normal_cdf(p, tolerance=tolerance)

    left, right = -10.0, 10.0
    while normal_cdf(right) - normal_cdf(left) > tolerance:
        mid = (right + left) / 2          # ì¤‘ì•™ê°’
        if normal_cdf(mid) < p:           # ì¤‘ì•™ê°’ì´ ì•„ì§ ì‘ìœ¼ë©´ ìœ—ìª½ì„ íƒìƒ‰
            left = mid
        elif normal_cdf(mid) > p:         # ì¤‘ì•™ê°’ì´ ì•„ì§ í¬ë©´, ì•„ë«ìª½ì„ íƒìƒ‰
            right = mid
        else:
            break
    return (right + left) / 2
````

* $X_1, \cdots, X_n$ ì„ í‰ê·  $\mu$, í‘œì¤€ í¸ì°¨ $\sigma$ë¥¼ ê°€ì§ˆ ë•Œ, $n$ì´ ì¶©ë¶„íˆ í¬ë©´, ë‹¤ìŒì€ í‰ê·  0, í‘œì¤€í¸ì°¨ 1ì¸ ì •ê·œ ë¶„í¬ì— ê·¼ì‚¬í•¨.
$$\frac{ X_1 + \cdots + X_n - n\mu}{\sigma \sqrt{n}}$$


#### ë² ë¥´ëˆ„ì´ ë¶„í¬ì™€ ì´í•­ ë¶„í¬

* ë² ë¥´ëˆ„ì´ ë¶„í¬ : pì˜ í™•ë¥ ë¡œ ì„±ê³µ ì‹œ 1, ì‹¤íŒ¨ ì‹œ 0
* ì´í•­ ë¶„í¬ : níšŒ ë…ë¦½ ë² ë¥´ëˆ„ì´ ì‹œí–‰ ì‹œ ì´ ì„±ê³µ íšŸìˆ˜

````python
def bernoulli_trial(p):
    return 1 if random.random() < p else 0
````

````python
def binomial(n, p):
    return sum(bernoulli_trial(p) for _ in range(n))
````

ì´í•­ ë¶„í¬ì˜ í‰ê·  : $np$,  í‘œì¤€í¸ì°¨ : $ \sqrt{np(1-p)}$

#### ì´í•­ë¶„í¬ì™€ ì¤‘ì‹¬ ê·¹í•œ ì •ë¦¬

````python
from collections import Counter
import random
def make_hist(n, p, num_points):
    data = [binomial(n, p) for _ in range(num_points)]
    # ì´í•­ ë¶„í¬ ìƒ˜í”Œì„ íˆìŠ¤í† ê·¸ë¨ìœ¼ë¡œ
    histogram = Counter(data)                  #from collections import Counter
    plt.bar([x for x in histogram.keys()],
            [v / num_points for v in histogram.values()], 0.8, facecolor='none',  edgecolor='black')
    
    mu, sigma = p * n, math.sqrt(n * p * (1 - p))
    # ì •ê·œ ë¶„í¬ ê·¸ë¦¬ê¸°
    xs = range(min(data), max(data) + 1)
    ys = [normal_pdf(i, mu, sigma) for i in xs]
    plt.plot(xs,ys)
    plt.show()
````

### ì •ê·œ ëª¨ì§‘ë‹¨ ì†Œí‘œë³¸

````python
import numpy as np
import matplotlib.pyplot as plt

N = 10000
mean_x = np.zeros(N)
std_x = np.zeros(N)
ts = np.zeros(N)

mu, sigma, n = 5, 3, 10      #ëª¨í‰ê· , ëª¨í‘œì¤€í¸ì°¨, í‘œë³¸ ê°œìˆ˜

for i in range(N):
    x = np.random.normal(mu, sigma, n)
    mean_x[i], std_x[i] = x.mean(), x.std(ddof=1)          # í‘œë³¸ í‰ê· , í‘œë³¸ í‘œì¤€í¸ì°¨
    ts[i] = (mean_x[i] - mu)/(std_x[i]/np.sqrt(n))   # t í†µê³„ëŸ‰
    
plt.hist(ts, bins=20, density=True)
plt.show()
````

### Random walks

ì¼ë ¨ì˜ ë¬´ì‘ìœ„ ë‹¨ê³„ë¡œ êµ¬ì„±ëœ ê²½ë¡œë¥¼ ì„¤ëª…í•˜ëŠ” ëœë¤ í™•ë¥ ê³¼ì •

````python
import numpy as np
import matplotlib.pyplot as plt

position = 0
walk = [position]
steps = 1000
for _ in range(steps):
    step = np.random.normal(loc=0, scale=1)
    position += step
    walk.append(position)
    
plt.plot(range(steps+1), walk) 
plt.show()
````

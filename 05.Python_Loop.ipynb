## ```while```

같은 코드를 ```while```문을 이용하여 여러 번 실행할 수 있음  
```while``` 루프 문은 주어진 조건이 참일 경우 목표 문장을 반복적으로 실행

간단한 예제부터 살펴보자.

````python
i = 0
while i < 5:
    print("Print this sentence repeatedly.")
    i += 1
````

````python
# by 22010726
x = 70
ans = 0
increment = 1
while increment >= 0.0001:
    ans += increment
    if ans ** 2 == x:
        break
    elif ans ** 2 > x:
        ans -= increment
        increment = increment / 10.0
print(ans)
````

다음은 보다 복잡한 예제로서 5를 3번 합하여 15를 만드는 과정을 보여준다.  즉, 5 * 3을 더하기(5 + 5 + 5)로 반복하는 과정.

````python
x = 0
j = 3
while j > 0:
    x += 5
    j -= 1
print(x)
````

### while문 문법

```
while < test statement >:  
   < test 결과가 True일 경우 해야 할 일 >
```
    
```if```문과 비슷한 구조를 지님   
* < test statement >가 ```True```일 경우 해야 할 일을 수행하고 다시 ```while```문으로 돌아가 test statement가 참인지 체크  
* < test statement >가 ```False```일 경우 ```while```문 종료  
* ```if``` 문과 마찬가지로 들여쓰기 규칙을 잘 지켜야 한다.

<div class="alert alert-block alert-warning"> <b>주의</b> <code>while</code>문 내의 코드를 잘 작성하여 의도하지 않은 무한 루프(infinite loop)가 발생하지 않도록 주의한다. 의도하지 않은 무한 루프는 프로그램 상의 오류는 아니지만 프로그램이 멈추지 않아 리소스를 소모하고 컴퓨터를 멈추게 하는 등 커널을 정상적으로 이용할 수 없게 한다..</div>

### 반복문의 일반적 형태

반복문을 설계하여야 할 때 다음 사항들을 생각해 보자.
* counter로 사용할 변수 설정 
* counter 변수를 반복문 바깥에서 초기화
* 반복문이 언제 끝나야 하는지 결정하는 end-test 설정
* 반복문 내에서 실행할 코드를 작성
  * counter를 변화시키는 작업 포함
* 반복문이 종료되었을 때 반복문의 결과를 이용해 무엇을 할지 작성

### 반복문 내의 반복문

````python
x = 3
ans = 0
while x > 0:
    y = 4
    x -= 1
    while y > 0:
        ans += 1
        y -= 1
print(ans)
````

### ```break```를 통한 ```while``` 벗어나기

```break```를 이용하면 현재의 loop를 벗어날 수 있음

(이후에 배울 ```for```문도 break를 통해 벗어날 수 있다.)

* 일반적으로 ```if```와 함께 사용되며 ```if```의 조건이 ```True```이면 현재 loop 종료

````python
count = 0
while True:
    count += 1
    if count >= 10:
        break
print(count)
````

### ```while/else```

Python에 있는 독특한 문법    
```while```문의 반복 조건이 ```False```일 경우 ```else```문을 실행
* ```while```문에 아예 진입하지 않았거나 정상적으로 종료될 경우 ```else```문 실행

```break```를 통해 ```while```문을 강제로 빠져나오면 ```else```문은 실행되지 않음  
(이후에 배울 ```for```문도 비슷하게 ```for/else```문을 적용할 수 있다.)

예제 : ```x```의 값을 바꿔가며 실험해 보자.

````python
x = 8
while x % 2 == 0:
    x = x / 2
    if x == 1:
        print("x is a power of 2")
        break
else:
    print("x is not a power of 2")
````

<p> &#9635 예제: 제곱근을 찾는 단순 반복문 프로그램 </p>

다음의 프로그램은 x가 제곱수일 경우 정수인 제곱근을 찾아준다.
* 단, 한 가지 잘못된 점이 있다. 찾아보자.

````python
x = 16
ans = 0
while ans * ans <= x:
    ans = ans + 1
print(ans)
````

위 프로그램은 정수의 제곱에만 한정된 프로그램이다.
한편, 다음은 양의 실수 범위에서 제곱근을 찾는 단순 반복문 프로그램이다.
* 아래의 프로그램이 어떻게 작동하는지 생각해 보자.
* 이 프로그램을 어떻게 하면 향상시킬 수 있는지 생각해 보자.
  * 여유가 되면 Bisection 방법에 대해 조사하여 보자.

````python
x = 18
ans = 0
increment = 1
while increment >= 0.0001:
    ans += increment
    if ans ** 2 == x:
        break
    elif ans ** 2 > x:
        ans -= increment
        increment = increment / 10.0
print(ans)
````

* 메르센 수(Mersenne number)는 2의 거듭제곱에서 1이 모자란 숫자를 가리킨다.
   * 즉 $2^n-1$의 형태로 나타난다.
   * $3 = 2^2-1$ 이므로 메르센 수이다.
* 1부터 1000까지의 숫자 중 메르센 숫자를 모두 찾는 프로그램을 작성하라.

````python
for i in range(1,10):
    print(2 ** i - 1)
````

* 메르센 소수는 메르센 수이면서 소수인 숫자를 나타낸다. 1부터 1000까지의 숫자 중 메르센 소수를 모두 찾는 프로그램을 작성하라.

````python
prime = []
mersenne = []
result = []

for i in range(1,10):
    mersenne.append(2 ** i - 1)
    
for i in range(2, 1001):
    count = 0
    for j in range(2, i):
        if i % j == 0:
            count += 1
    if count == 0:
        prime.append(i)

result = list(set(prime).intersection(mersenne))
print(result)
````

* 거꾸로 해도 똑같은 숫자를 palindrom이라 한다. 1부터 200까지의 숫자 중 palindrom이면서 소수인 수를 모두 찾는 프로그램을 작성하라.

````python
prime = []
palindrom = []
result = []
for i in range(1, 201):
    x = str(i)
    if x[0] == x[-1]:
        palindrom.append(int(x))
    
    count = 0
    for j in range(2, i):
        if i % j == 0:
            count += 1
    if count == 0:
        prime.append(i)

for i in range(len(prime)):
    for j in range(len(palindrom)):
        if prime[i] == palindrom[j]:
            result.append(prime[i])

result.remove(1)
print(result)
````

* 임의의 자연수를 선택하여, 만약 짝수이면 2로 나누고, 홀수이면 3을 곱하고 1을 더하라. 이 과정을 반복하면 항상 1에 도달하는지 체크해 보라.

````python
x = int(input("숫자 입력: "))
while x != 1:
    if x % 2 == 0:
        x = x / 2
        print(x)
    else:
        x = 3 * x + 1
        print(x)
````
